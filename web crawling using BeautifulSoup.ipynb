```python
from bs4 import BeautifulSoup
import requests

def get_html(url):  # 페이지에 소스코드를 요청하여 받는 함수로
    _html = ''      # 앞으로 변경하지 않고 사용하시면 됩니다.
    resp = requests.get(url)
    if resp.status_code == 200:
        _html = resp.text
    return _html
```

```python
# 원하는 페이지의 URL을 아래 String 형식으로 입력하시면,
URL = 'https://finance.naver.com/sise/lastsearch2.nhn'

# 위에서 만든 함수를 통해 해당 주소의 소스코드를 받고
html = get_html(URL)

# BeautifulSoup 라이브러리의 기능으로 html형식으로 parsing합니다.
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# 아래와 같이 원하는 영역의 HTML 문법 규칙 입력
# <table cellpadding="0" cellspacing="0" class="type_5">
Code_text = soup('table',{'cellpadding' : "0", 'cellspacing' : "0"})
print(Code_text)
```

```python
# Code_text는 Soup형태의 변수이므로 String형태로 변형
Str_Code = str(Code_text)

# 원하는 규칙으로 split 진행후 List에 저장
Split_list = Str_Code.split('<a class="tltle" href="/item/main.nhn?code=')

# 우리가 원하는 키워드는 규칙 뒤에 위치하기 때문에 1번 Index부터 출력
for i,ii in enumerate(Split_list[1:]):
    print(i,'번째 구분자 뒤 문자열 => ',ii)
```

```python
# Code_text는 Soup형태의 변수이므로 String형태로 변형
Str_Code = str(Code_text)

# 원하는 규칙으로 split 진행후 List에 저장
Split_list = Str_Code.split('<a class="tltle" href="/item/main.nhn?code=')

# 한번 Split이 적용된 문자열(ii)에 "> 구분자로 다시 Split함수 적용
for i,ii in enumerate(Split_list[1:]):
    # Rule1 : "> 바로 뒤(Split을 적용한 후 1번 index에 해당하는 값)에 나오는 Text 저장
    Str_Rule1 = ii.split('">')[1]
    # Rule2 : Rule1을 적용한 Text에 </a> 의 바로 앞(Split 적용 후 0번 index)에 나오는 Text 저장
    Str_Rule2 = Str_Rule1.split('</a>')[0]
    print('맨 처음 Split 된 문자열', i,'에서 Rule1,2적용 결과 => ',Str_Rule2)
```
